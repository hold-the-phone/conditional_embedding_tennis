```python
import tensorflow as tf
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder,StandardScaler

pd.options.display.max_columns=150
```

```python
df = pd.read_csv('mini_tennis.csv')
#print(df.columns)
df = df[['Location1', 'Tournament', 'Date', 'Court', 'Surface', 'Round',
       'Best of', 'Player1', 'Player2', 'P1Rank', 'P2Rank', 'P1Pts', 'P2Pts',
       'P1_1', 'P2_1', 'P1_2', 'P2_2', 'P1_3', 'P2_3', 'P1_4', 'P2_4', 'P1_5',
       'P2_5', 'P1_sets', 'P2_sets','year', 'month', 'set_win_pct', 'P1_games', 
       'P2_games', 'games_count','game_win_pct', 'p1_id', 'p2_id', 'location_id', 
       'month_id', 'round_id','indoor', 'outdoor', 'grass', 'clay', 'hard', 'carpet', 
       'bestof3','bestof5', 'match_last_night', 'games_last_night', 'match_win']]

print(df.shape)
df.dropna(inplace=True)
print(df.shape)
le = LabelEncoder()
df['p1_id'] = le.fit_transform(df['Player1'])
df['p2_id'] = le.transform(df['Player2'])
#print(df.columns)
df['Date'] = pd.to_datetime(df['Date'])

df = df[(df['PS_P1']>=1.3)&(df['PS_P2']>=1.3)]

print(df.shape)

df['P1Rank'] = df['P1Rank'].astype(float)
df['P2Rank'] = df['P2Rank'].astype(float)

df
```

```python
df_train = df[(df['Date']<pd.to_datetime('2018-01-01'))&
                   (df['Date']>pd.to_datetime('2016-01-01'))]
df_train.reset_index(inplace=True)

df_test = df[df['Date']>=pd.to_datetime('2018-01-01')]
df_test.reset_index(inplace=True)
```

```python
np.max(df['p2_id'])+1
```

#### tf model

```python
tf.reset_default_graph()

DICTIONARY_LENGTH = np.max(df['p2_id'])+1

EMBEDDING_SIZE=10

p1_set = ['P2Rank','indoor', 'grass', 'clay', 'hard',
       'bestof5', 'match_last_night', 'games_last_night']

p2_set = ['P1Rank','indoor', 'grass', 'clay', 'hard', 'carpet', 'bestof5']

y = tf.placeholder(tf.float32,shape=[None,1])
x1 = tf.placeholder(tf.float32,shape=[None,8])
x2 = tf.placeholder(tf.float32,shape=[None,7])

idx_1 = tf.placeholder(tf.int32,shape=[None,1])
idx_2 = tf.placeholder(tf.int32,shape=[None,1])

b_initializer = tf.zeros_initializer()
w_initializer = tf.initializers.random_normal(stddev=0.2)


###CONDITIONAL EMBEDDING #1

w1 = tf.get_variable(name='w1',shape=[DICTIONARY_LENGTH,8,EMBEDDING_SIZE],initializer=w_initializer)

b1 = tf.get_variable("b1",dtype=tf.float32,shape=[DICTIONARY_LENGTH],initializer=b_initializer)

correct_weight_mat1 = tf.reshape(w1[idx_1,:,:],(8,EMBEDDING_SIZE))

correct_bias1 = tf.reshape(b1[idx_1],())

activation1 = tf.nn.relu((tf.matmul(x1,correct_weight_mat1)+correct_bias1))

embbed1 = tf.layers.batch_normalization(activation1)


###CONDITIONAL EMBEDDING #2

w2 = tf.get_variable(name='w2',shape=[DICTIONARY_LENGTH,7,EMBEDDING_SIZE],initializer=w_initializer)

b2 = tf.get_variable("b2",dtype=tf.float32,shape=[DICTIONARY_LENGTH],initializer=b_initializer)

correct_weight_mat2 = tf.reshape(w2[idx_2,:,:],(7,EMBEDDING_SIZE))
correct_bias2 = tf.reshape(b2[idx_2],())

activation2 = tf.nn.relu((tf.matmul(x2,correct_weight_mat2)+correct_bias2))

embbed2 = tf.layers.batch_normalization(activation2)


###Rest of the Graph and Train and Output (nothing new)

joined = tf.concat([embbed1,embbed2],axis=1)

h2 = tf.layers.dense(inputs=joined,units=128,activation=tf.nn.relu,
                      kernel_initializer=w_initializer,
                      bias_initializer=b_initializer)

output = tf.layers.dense(inputs=h2,units=1,activation=None,
                      kernel_initializer=w_initializer,
                      bias_initializer=b_initializer)

loss = tf.losses.mean_squared_error(y,output)

trainstep = tf.train.AdamOptimizer().minimize(loss)

init = tf.global_variables_initializer()
  
sess = tf.Session()
sess.run(init)


train_x1 = np.array(df_train[p1_set],dtype=np.float32)
#train_x1 = train_x1.reshape((-1,1))
train_x2 = np.array(df_train[p2_set],dtype=np.float32)
#train_x2 = train_x2.reshape((-1,1))

train_idx1 = np.array(df_train['p1_id'],dtype=np.float32)
train_idx1 = train_idx1.reshape((-1,1))
train_idx2 = np.array(df_train['p2_id'],dtype=np.float32)
train_idx2 = train_idx2.reshape((-1,1))

train_y = np.array(df_train['set_win_pct'],dtype=np.float32)
train_y = train_y.reshape((-1,1))

test_x1 = np.array(df_test[p1_set],dtype=np.float32)
#test_x1 = test_x1.reshape((-1,1))
test_x2 = np.array(df_test[p2_set],dtype=np.float32)
#test_x1 = test_x1.reshape((-1,1))

test_idx1 = np.array(df_train['p1_id'],dtype=np.float32)
test_idx1 = test_idx1.reshape((-1,1))
test_idx2 = np.array(df_train['p2_id'],dtype=np.float32)
test_idx2 = test_idx2.reshape((-1,1))

test_y = np.array(df_test['set_win_pct'],dtype=np.float32)
test_y = test_y.reshape((-1,1))

with tf.train.MonitoredSession() as sess:
    for i in range(500):
        sess.run(trainstep,feed_dict={x1:train_x1,x2:train_x2,idx_1:train_idx1,idx_2:train_idx2,y:train_y})
        if i%10==0:
            loss_train = sess.run(loss,feed_dict={x1:train_x1,x2:train_x2,idx_1:train_idx1,idx_2:train_idx2,y:train_y})
            loss_test = sess.run(loss,feed_dict={x1:test_x1,x2:test_x2,idx_1:test_idx1,idx_2:test_idx2,y:test_y})
            print('\ntrain accuracy is:' + str(loss_train))
            print('test accuracy is:' + str(loss_test)+'\n')
            
    a = sess.run(output,feed_dict={x1:test_x1,x2:test_x2,idx_1:test_idx1,idx_2:test_idx2,y:test_y})
    print(a)

```
